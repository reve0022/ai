```python
#(영어)알파벳과 숫자를 섞어서 입력을 받아서 숫자와 문자를
# 분리하여 출력하는 프로그램을 작성하시오.

#[입력]
#숫자를 포함한 문자열을 입력하세요 >>> 123abc45
#[출력]
#문자: abc
#숫자: 12345

str = input('숫자를 포함한 문자열을 입력하세요 >>> ')

al=''
no=''


for i in str:
  if i.isalpha():
    al = al + i
  elif i.isdigit():
    no = no + i

print('문자:',al)
print('숫자:',no)
```

```python
#A답변
# by 이지은

str = input('숫자를 포함한 문자열을 입력하세요 >>> ')
cnt1 = 0
al = ''
cnt2 = 0
no = ''

for i in str:
  if i == '1' or i == '2' or i == '3' or i == '4' or i == '5' or i == '6' or i == '7' or i == '8' or i == '9' or i == '0':
    cnt2+=2
    no+=i
  else:
    cnt1+=1
    al+=i

print(f'문자({cnt1}):{al}')
print(f'숫자({cnt2}):{no}')
```

```python
#A답변
def str_slide( pr ):
    s = "1234567890"
    a = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    str = "";
    int = "";
    count_int = 0; 
    count_str = 0;
    
    for i in pr:
        if i in s:
            int += f"{i} ";
            count_int +=1 ;
        elif i in a:
            str += f"{i} ";
            count_str += 1;            
            
    print(f"입력 숫자 : {int} {count_int}개");
    print(f"입력 문자 : {str} {count_str}개");
            
pr = input("입력 > ");
str_slide(pr);
```

```python
#A+ 답변

# by 민지수

str=input('숫자를 포함한 문자열을 입력하세요>>> ')

ai=""
no=""

for i in str :
    if 64<ord(i)<123:
        ai=ai+i
    
    elif 48<ord(i)<58:
        no=no+i
    
    
print(f'문자({len(ai)}) : ',ai)
print(f'숫자({len(no)}): ',no)
```

```python
# 정규 표현식으로 고쳤습니다.
import re;
def str_slide( pr ):
    n = re.compile("[0-9]");
    a = re.compile("[a-zA-Z]");
    str = "";
    int = "";
    count_int = 0; 
    count_str = 0;
    
    for i in pr:
        if a.search(i):
            int += f"{i} ";
            count_int +=1 ;
        elif n.search(i):
            str += f"{i} ";
            count_str += 1;            
            
    print(f"입력 숫자 : {int} {count_int}개");
    print(f"입력 문자 : {str} {count_str}개");
            
pr = input("입력 > ");
str_slide(pr);
```

```python
# 이진 탐색법 Binary Serch

# 탐색 대상 데이터들이 미리 오름차순 또는 내림차순으로 정렬이 되어있어야만 한다.
# 탐색하는 범위를 절반씩 추려나가면서 탐색을 한다.

0   1   2   3   4   5   6   index
11 13  17  19  23  19  31   data 


0  1  2  3  4  5  6  step 1 가운데 인덱스 즉, 3을 열어서 확인한다.
         19                 찾고 있는 17이 아니다. 그런데 정렬되어 있다는 전제조건이 있으므로
                            이후의 모든 데이터는 절대 찾고있는 17이 될수 없다.
0  1  2              step 2 새로운 범위의 가운데 인덱스 즉 1을 열어 확인한다.
   13                        찾고 있는 17이 아니다. 그런데 정렬되어 있다는 전제조건이 있으므로
                            이후의 모든 데이터는 절대 찾고있는 17이 될수 없다.
2                    step 3 마지막 요소만 남게 되었지만 찾고 있는 17이 아예 없는 경우도 
17                          발생할수 있으므로 처리해 주어야 하지만 지금은 마지막 index 2 가
                            찾고 있는 17이므로 검색을 종료한다.
```

```python
# 이진 탐색법의 알고리즘
# 1. 가운데 요소를 선택하는 처리(인덱스의 평균을 계산하여 구한다.)
# 2. 가운데 데이터와 찾는 데이터를 비교하는 처리
# 3. 탐색 범위를 절반으로 줄이는 처리

# 요소의 갯수가 짝수 일때는 어떻게 해야할까?

#    0   1   2   3   4   5  
#    11  13  17  19  23  29
#            --  --
# 요소의 갯수가 짝수 일때는 가운데 요소가 2개라는 것을 알수 있따.
# 이 경우 어느쪽을 선택해야 한다고 정해진 것은 없다. 단지 둘중 하나만 선택하면 된다.
# 따라서 같은 공식으로 계산해 보년
# (0+5)/2 = 2.5
# 소수를 정수화 하는 방법으로는 1. 반올림  2.올림  3.버림 이있다.
# 하지만 제일 간단한 방법은 3. 버림 이 제일 쉬운 처리이다.
# 정수형 변수에 넣기만 하면 자동으로 소수 부분은 잘려 나가기 때문에
# 쉽게 정수 부분만 얻을수 있게 된다.

```
