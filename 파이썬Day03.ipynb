```python
# while 중첩
# while  내부에 또 다른  while 을 사용하는 것을 중첩이라고 한다.

# 3 일 동안 1~3교시 수업을 한다.

# 1일차 1교시입니다.
# 1일차 2교시입니다
# 1일차 3교시입니다.
# 2일차 1교시입니다.
# .
# .
# .
# 3일차 3교시입니다.

day = 1
while day <= 3:
  hour = 1    #중요 어디서 초기화를 하느냐
  while hour <= 3:
    print(f'{day}일차 {hour}교시입니다.')
    hour += 1
  day += 1
```

```python
# 커피 1잔을 300원에 판매하는 자판기가 있다 
# 자판기에 돈을 넣었을때 자팜기에서 뽑을 수 있는 커피가 몇잔이고
# 잔돈이 얼마인지 함께 출력하세요
# 자판기에 얼마를 넣을까요 >>> 1400
#커피 1잔, 잔돈 1100원
#커피 2잔, 잔돈 800원
#커피 3잔, 잔돈 500원
#커피 4잔, 잔돈 200원

a=int(input('자판기에 얼마를 넣을까요>>>'))
cof=300
num=1
while cof<=a:
  print(f'커피 {num}잔, 잔돈 {a-cof}원')
  num+=1
  cof+=300
```

```python
# 비밀 번호(str)를 입력 받아서 숫자와 문자가 모두 반드시 하나이상 
# 포함된 경우만 '사용 가능한 비번입니다.'
# '불가능한 비번입니다'
# ch.isalpha() - 알파벳이면 구성되면 True
# ch.isnumeric() - 숫자면 True

# (입력) 비밀 번호를 입력하세요 >>>
# (출력) 사용 가능한 비번입니다.
pw = input('비밀 번호를 입력하세요 >>>')
ch_count = 0
no_count = 0

for ch in pw:
  if ch.isalpha():
    ch_count += 1
  elif ch.isnumeric():
    no_count += 1

if ch_count > 0 and no_count > 0:
  print('사용 가능한 비번입니다.')
else:
  print('사용 불가한 비번입니다.')
```

```python
#사용자로부터 양의 정수를 하나 입력 받아서 1부터 입력 받은 정수까지의
# 모든 정수의 합계를 구하는 프로그램을 작성하세요
#(입력) 임의의 양수를 입력하세요>>>>5
#(출력) 1부터 5 사이의 모든 정수의 합계는 15입니다.


no = int(input('임의의 양수를 입력하세요 >>>'))

sum = 0

for i in range(1,no+1):
  sum = sum + i  # sum += i

print(f'1부터 {no}사이의 모든 정수의 합계는 {sum}')
```

```python
#사용자로부터 임의의 정수를 하나 입력 받아서 그 숫자 만큼의 '과일이름'입력받아서
#'basket' 리스트에 저장하는 프로그램을 작성하세요.
# 몇 개의 과일을 보관할까요?>>>>>>3
#1번째 과일을 입력하세요>>>>사과
#2번째 과일을 입력하세요>>>>바나나
#3번째 과일을 입력하세요>>>>체리
#입력 받은 과일들은 ['사과','바나나','체리']입니다.

no = int(input('몇 개의 과일을 보관할까요 ? >>>'))

basket = []

for i in range(1, no+1):
  fruit = input(f'{i}번째 과일을 입력하세요 >>>')
  basket.append(fruit)

print(f'입력 받은 과일들은 {basket} 입니다.')
```
